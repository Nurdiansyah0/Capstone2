# Data Kamus Pelanggan Supermarket

- Pelanggan
  - ID: Identifikasi unik pelanggan
  - Year_Birth: Tahun lahir pelanggan
  - Education: Tingkat pendidikan pelanggan
  - Marital_Status: Status perkawinan pelanggan
  - Income: Pendapatan rumah tangga tahunan pelanggan
  - Kidhome: Jumlah anak di rumah tangga pelanggan
  - Teenhome: Jumlah remaja di rumah tangga pelanggan
  - Dt_Customer: Tanggal pendaftaran pelanggan dengan perusahaan
  - Recency: Jumlah hari sejak pembelian terakhir pelanggan
  - Complain: 1 jika pelanggan mengeluh dalam 2 tahun terakhir, 0 jika tidak   
- Produk
  - MntWines: Jumlah pengeluaran untuk anggur dalam 2 tahun terakhir
  - MntFruits: Jumlah pengeluaran untuk buah-buahan dalam 2 tahun terakhir
  - MntMeatProducts: Jumlah pengeluaran untuk daging dalam 2 tahun terakhir
  - MntFishProducts: Jumlah pengeluaran untuk ikan dalam 2 tahun terakhir
  - MntSweetProducts: Jumlah pengeluaran untuk makanan manis dalam 2 tahun terakhir
  - MntGoldProds: Jumlah pengeluaran untuk produk emas dalam 2 tahun terakhir
- Promosi
  - NumDealsPurchases: Jumlah pembelian yang dilakukan dengan diskon
  - AcceptedCmp1: 1 jika pelanggan menerima penawaran dalam kampanye ke-1, 0 jika tidak
  - AcceptedCmp2: 1 jika pelanggan menerima penawaran dalam kampanye ke-2, 0 jika tidak
  - AcceptedCmp3: 1 jika pelanggan menerima penawaran dalam kampanye ke-3, 0 jika tidak
  - AcceptedCmp4: 1 jika pelanggan menerima penawaran dalam kampanye ke-4, 0 jika tidak
  - AcceptedCmp5: 1 jika pelanggan menerima penawaran dalam kampanye ke-5, 0 jika tidak
  - Response: 1 jika pelanggan menerima penawaran dalam kampanye terakhir, 0 jika tidak
- Tempat
  - NumWebPurchases: Jumlah pembelian yang dilakukan melalui situs web perusahaan
  - NumCatalogPurchases: Jumlah pembelian yang dilakukan menggunakan katalog
  - NumStorePurchases: Jumlah pembelian yang dilakukan langsung di toko
  - NumWebVisitsMonth: Jumlah kunjungan ke situs web perusahaan dalam bulan terakhir

# Data Cleansing
<font size="3"> Dari data yang ada, data yang missing hanya ada *total 24 data* pada tabel income, meskipun data yang hilang hanya *1%* ,untuk meningkatkan akurasi pada prescriptive dan memperkecil outliers pada preditive analisis, missing data akan di lakukan cleaning menggunakan nilai dari rata-rata penghasilan.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.impute import SimpleImputer
# Membuat objek SimpleImputer dengan strategi mean
imputer = SimpleImputer(strategy='mean')
# Memasukkan nilai ke dalam kolom 'income'
df['Income'] = imputer.fit_transform(df[['Income']])
```

```python
# Memeriksa missing Value kembali
missing = df.isna().sum()
print(missing)
```

```python
df = df.select_dtypes(include = ['float64', 'int64'])
```

```python
print("Skewness: \n", list(zip(df.skew().index,df.skew().values)))
```

```python
# Calculate skewness
skewness_data = list(zip(df.skew().index, df.skew().values))

# Extract column names and skewness values
column_names = [col[0] for col in skewness_data]
skewness_values = [col[1] for col in skewness_data]

# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.bar(column_names, skewness_values)
plt.xlabel('Columns')
plt.ylabel('Skewness')
plt.title('Skewness of Numerical Columns')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(True)
plt.show()

# Print skewness values
print("Skewness: \n", skewness_data)
```

```python

def calculate_iqr(data):
  """
  Calculates the Interquartile Range (IQR) dataset.

  Args:
    data: A list or NumPy array containing the data.

  Returns:
    The Interquartile Range (IQR) of the data.
  """

  q1 = np.percentile(data, 25)  # Calculate the first quartile (25th percentile)
  q3 = np.percentile(data, 75)  # Calculate the third quartile (75th percentile)
  iqr = q3 - q1  # Calculate the IQR

  return iqr

# Example usage:
iqr_value = calculate_iqr(df)
print(f"IQR: {iqr_value}")
```
